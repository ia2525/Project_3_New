Repeating the above process without pandas. I created different functions for mean, median, and mode, all using lists or dictionaries in order to store values before sorting them and computing the output.

They follow a similar process of opening the file in read mode, processing the the rows of the excel via DictReader and adding all values to the list/dictionary

```python
import csv

def mean_dbh(data, column_name):
    values = []
    with open(data, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            if row[column_name]:
                values.append(float(row[column_name]))

    values.sort() #It was necessary to sort the values within the list, as without it my returned mean varied

    total = sum(values)
    count = len(values)
    return total / count

def median_dbh(data, column_name):
    values = []
    with open(data, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            if row[column_name]:
                values.append(float(row[column_name]))

    sorted_values = sorted(values) #here, it is necessary to sort the values in order to calculate median

    n = len(sorted_values)
    median = (sorted_values[n // 2] + sorted_values[(n - 1) // 2]) / 2 if n % 2 == 0 else sorted_values[n // 2]
    return median #Used gemini to assist with code above defining how to calculate median from the sorted values

def mode_dbh(data, column_name):
    counts = {} #because we are interested in frequency of values now, I used a dictionary to access the counts of each value directly
    with open(data, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            if row[column_name]:
                value = row[column_name]
                if value in counts:
                    counts[value] += 1 #code differentiates new values from additional values already in the dictionary
                else:
                    counts[value] = 1

        mode = float(max(counts, key=counts.get) if counts else None) #Used Gemini to assist with a code for determining the value with the highest count
    return mode

data = '/content/2015_Street_Tree_Census_-_Tree_Data_20241108.csv'
column_name = 'tree_dbh'
print(f"The mean tree_dbh is: {mean_dbh('/content/2015_Street_Tree_Census_-_Tree_Data_20241108.csv', 'tree_dbh'):.2f}")
print(f"The median tree_dbh is: {median_dbh('/content/2015_Street_Tree_Census_-_Tree_Data_20241108.csv', 'tree_dbh'):.2f}")
print(f"The mode tree_dbh is: {mode_dbh('/content/2015_Street_Tree_Census_-_Tree_Data_20241108.csv', 'tree_dbh'):.2f}")
```

```python
import csv
import math

data_file = '/content/2015_Street_Tree_Census_-_Tree_Data_20241108.csv'

tree_dbh_values = []
with open(data_file, 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        if row['tree_dbh']:
            tree_dbh_values.append(float(row['tree_dbh']))

bin_size = 15 #compacting the number of x values by increasing width of each segment

frequency = {}
for value in tree_dbh_values:
    bin_index = math.floor(value / bin_size) #used gemini to assist with creating a frequency of these values
    frequency[bin_index] = frequency.get(bin_index, 0) + 1

sorted_frequency = dict(sorted(frequency.items()))

asterisk_worth = 5000  # I had too many asterisks, so I changed their value

for bin_index, count in sorted_frequency.items():
    bar_height = math.ceil(count / asterisk_worth)  #used Gemini to assist with bringing together bigger bin size & asterisk worth into appropriate bar height
    bar = '*' * bar_height
    print(f"{bin_index * bin_size:3}-{bin_index * bin_size + bin_size:3}: {bar}") #Used Gemini to assist with this format
```
